```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation",model="gpt2")
```

```python
def generate_fake_news(prompt):
  full_prompt = f"Fake News Article:\n{prompt}\n"
  output = generator(full_prompt, max_new_tokens=400,do_sample=True, temperature=0.9, top_k=50, top_p=0.95)
  return output[0]['generated_text']
```

```python
user = input("Enter the content for fake news: ")
print(generate_fake_news(user))
```

# ***Detector***

```python
classifier = pipeline("text-classification", model="jy46604790/Fake-News-Bert-Detect",tokenizer="jy46604790/Fake-News-Bert-Detect")
```

```python
def detect_fake_news(news):
  result = classifier(news)[0]
  label = result['label']
  score = round(result['score']*100,2)

  if label.upper() == "LABEL_0":
    return f"Fake News Detected ({score}% confidence)"
  else:
    return f"Real News Detected ({score}% confidence)"
```

```python
user = input("Enter the news/article: ")
print(detect_fake_news(user))
```
